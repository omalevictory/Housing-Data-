```python
plt.figure(figsize=[20,7])
axis=sns.countplot(x=bo)
unique, freq = np.unique(bo, return_counts=True)
average = np.mean(freq)
plt.axhline(average)
plt.annotate('average',[2,average], xytext=[3,average + 1500],arrowprops={'facecolor':'blue'},fontsize=15)
plt.title('THE AVERAGE CRIMES COMMITED BASED ON BORO',fontdict = {'fontsize':14,'color':'black','fontweight': 'bold'})
```
